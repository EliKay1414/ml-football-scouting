```python
# %% [markdown]
# # Football Player Scouting Analysis
# 
# This notebook performs comprehensive cleaning and analysis of FIFA player datasets from 2015-2022.

# %%
# 1. Initial Setup
import sys
from pathlib import Path
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns
import json
from IPython.display import display, Markdown
    

# Set up visualization style
plt.style.use('seaborn-v0_8')
sns.set_palette('colorblind')
plt.rcParams['figure.figsize'] = (12, 6)
%matplotlib inline

# %%
# 2. Project Configuration
PROJECT_ROOT = Path(r"C:\Users\Nanaba\Desktop\football_player_scouting_ml")
SRC_DIR = PROJECT_ROOT / "src"
DATA_DIR = PROJECT_ROOT / "data"
VIS_DIR = PROJECT_ROOT / "visualizations"
REPORTS_DIR = PROJECT_ROOT / "reports"

# Create directories
for dir_path in [VIS_DIR, REPORTS_DIR, DATA_DIR / "processed"]:
    dir_path.mkdir(parents=True, exist_ok=True)

# Add src to Python path
if str(SRC_DIR) not in sys.path:
    sys.path.insert(0, str(SRC_DIR))

print(f"Project root: {PROJECT_ROOT}")

# %%
# 3. Data Loading
try:
    # Import the cleaner
    from data_cleaner import clean_fifa_data
    print(" Data cleaner imported successfully")
    
    # Load raw data
    raw_path = DATA_DIR / "raw" / "Career Mode player datasets - FIFA 15-22.xlsx"
    print(f"\nLoading data from: {raw_path}")
    
    # Read all sheets and add year column
    all_sheets = pd.read_excel(raw_path, sheet_name=None, engine='openpyxl')
    dfs = []
    
    for sheet_name, df in all_sheets.items():
        year_match = re.search(r'(\d{2})', sheet_name)
        if year_match:
            year = 2000 + int(year_match.group(1))
            df['year'] = year
            dfs.append(df)
    
    raw_df = pd.concat(dfs, ignore_index=True)
    print(f" Loaded {len(raw_df):,} raw records from {len(all_sheets)} sheets")
    
    # Show initial data sample
    display(Markdown("### Raw Data Sample (First 5 Rows)"))
    display(raw_df.head())
    
except Exception as e:
    print(f"\n Error during setup: {str(e)}")
    raise

# %%
# Show missing values heatmap
def plot_missing_values(df):
    """Visualize missing values pattern"""
    # Calculate missing percentage
    missing_pct = df.isna().mean().sort_values(ascending=False)
    
    # Plot top 20 columns with most missing values
    plt.figure(figsize=(15, 8))
    missing_pct.head(20).plot(kind='bar')
    plt.title("Top 20 Columns by Missing Value Percentage")
    plt.ylabel("Percentage Missing")
    plt.axhline(y=0.5, color='r', linestyle='--', 
                label='50% Threshold (Columns will be dropped)')
    plt.legend()
    plt.xticks(rotation=45, ha='right')
    plt.savefig(VIS_DIR / "missing_values_heatmap.png", dpi=300, bbox_inches='tight')
    plt.show()

# Run initial visualizations
print("\nVisualizing initial data distributions...")
plot_missing_values(raw_df)

# %%
# 5. Data Cleaning
from data_cleaner import FIFADataCleaner
print("\nStarting data cleaning process...")
try:
    cleaner = FIFADataCleaner(raw_df, PROJECT_ROOT)
    cleaned_df = cleaner.clean()
   
    
    # Save cleaning report
    report_path = REPORTS_DIR / "cleaning_report.json"
    with open(report_path, 'w') as f:
        json.dump(cleaner.cleaning_report, f, indent=2)
    
    print(f" Cleaning completed successfully! Report saved to {report_path}")
    
    # Show cleaning summary
    display(Markdown("### Cleaning Summary"))
    
    # Get detailed missing value handling report
    missing_data_report = cleaner.cleaning_report['steps'].get('missing_data', {})
    
    # Calculate total imputed values
    total_imputed = sum(
        v['count'] if 'count' in v else 0 
        for k, v in missing_data_report.items() 
        if not k.startswith('dropped_')
    )
    
    # Get dropped columns
    dropped_cols = [
        col.replace('dropped_', '') 
        for col in missing_data_report.keys() 
        if col.startswith('dropped_')
    ]
    
    summary_data = {
        "Metric": [
            "Initial Records", 
            "Final Records", 
            "Columns Removed", 
            "Duplicates Removed",
            "Columns Dropped (High Missing)",
            "Missing Values Imputed",
            "Remaining Missing Values"
        ],
        "Value": [
            len(raw_df),
            len(cleaned_df),
            len(raw_df.columns) - len(cleaned_df.columns),
            cleaner.cleaning_report['steps']['duplicate_handling']['duplicates_removed'],
            len(dropped_cols),
            total_imputed,
            cleaned_df.isna().sum().sum()  # Should be 0 with our new handler
        ]
    }
    
    display(pd.DataFrame(summary_data))
    
    if dropped_cols:
        display(Markdown(f"**Columns dropped (>50% missing):** {', '.join(dropped_cols)}"))
    
    # Show detailed missing value handling
    display(Markdown("### Missing Value Handling Details"))
    missing_details = []
    for col, stats in missing_data_report.items():
        if col.startswith('dropped_'):
            missing_details.append({
                "Column": col.replace('dropped_', ''),
                "Action": stats['action'],
                "Missing %": f"{stats['missing_pct']*100:.1f}%",
                "Imputed Value": "N/A (dropped)"
            })
        else:
            col_name = col.replace('_missing', '')
            missing_details.append({
                "Column": col_name,
                "Action": stats.get('action', 'unknown'),
                "Missing Count": stats.get('count', 'unknown'),
                "Imputed Value": str(stats.get('imputed_value', 'unknown'))
            })
    display(pd.DataFrame(missing_details))
    
    # Show cleaned data sample
    display(Markdown("### Cleaned Data Sample (First 5 Rows)"))
    display(cleaned_df.head())
    
except Exception as e:
    print(f"\n Error during cleaning: {str(e)}")
    raise
# %%
# 6. Post-Cleaning Analysis

# Show remaining missing values
def plot_remaining_missing(df):
    """Visualize remaining missing values after cleaning"""
    remaining_missing = df.isna().sum()
    remaining_missing = remaining_missing[remaining_missing > 0]
    
    if len(remaining_missing) > 0:
        plt.figure(figsize=(12, 6))
        sns.barplot(x=remaining_missing.index, y=remaining_missing.values)
        plt.title("Remaining Missing Values After Cleaning")
        plt.ylabel("Count")
        plt.xticks(rotation=45, ha='right')
        plt.savefig(VIS_DIR / "remaining_missing.png", dpi=300, bbox_inches='tight')
        plt.show()
    else:
        print(" No remaining missing values in the dataset")

print("\nVisualizing cleaned data distributions...")
plot_remaining_missing(cleaned_df)

# %%
# 7. Data Export and Final Summary
try:
    # Save cleaned data
    output_path = DATA_DIR / "processed" / "fifa_players_cleaned.csv"
    cleaned_df.to_csv(output_path, index=False)
    
    # Save as pickle for faster loading
    pickle_path = DATA_DIR / "processed" / "fifa_players_cleaned.pkl"
    cleaned_df.to_pickle(pickle_path)
    
    print(f"\n Saved cleaned data to:")
    print(f"- CSV: {output_path}")
    print(f"- Pickle: {pickle_path}")
    print(f"\nFinal dataset shape: {cleaned_df.shape}")
    
    # Show generated visualizations
    print("\nGenerated Visualizations:")
    for viz in sorted(VIS_DIR.glob("*.png")):
        print(f"- {viz.name}")
    
    # Show memory usage
    display(Markdown("### Memory Usage Summary"))
    mem_usage = cleaned_df.memory_usage(deep=True).to_frame(name='Bytes')
    mem_usage['MB'] = mem_usage['Bytes'] / (1024 ** 2)
    display(mem_usage.sort_values('MB', ascending=False).head(10))
    
    # Show data types
    display(Markdown("### Data Types Summary"))
    display(pd.DataFrame(cleaned_df.dtypes, columns=['Data Type']))
    
except Exception as e:
    print(f"\n Error during export: {str(e)}")
    raise

# %%
# 8. Show Cleaning Report Highlights
try:
    with open(REPORTS_DIR / "cleaning_report.json", 'r') as f:
        report = json.load(f)
    
    display(Markdown("## Data Cleaning Report Highlights"))
    
    # Initial vs Final
    display(Markdown("### Dataset Size Changes"))
    display(pd.DataFrame({
        "Metric": ["Initial Records", "Final Records", "Columns Removed"],
        "Value": [
            report['initial_columns']['total'],
            len(cleaned_df.columns),
            report['initial_columns']['total'] - len(cleaned_df.columns)
        ]
    }))
    
    # Missing data handling details
    if 'steps' in report and 'missing_data' in report['steps']:
        display(Markdown("### Missing Data Handling"))
        missing_stats = []
        for col, stats in report['steps']['missing_data'].items():
            if col.startswith('dropped_'):
                missing_stats.append({
                    "Column": col.replace('dropped_', ''),
                    "Action": stats['action'],
                    "Missing %": f"{stats['missing_pct']*100:.1f}%"
                })
            else:
                missing_stats.append({
                    "Column": col.replace('_missing', ''),
                    "Action": stats['action'],
                    "Missing Values": stats.get('count', 'N/A')

                })
        display(pd.DataFrame(missing_stats))
    
    # Issues found
    if 'issues' in report and report['issues']:
        display(Markdown("### Key Issues Found"))
        issues_df = pd.DataFrame.from_dict(report['issues'], orient='index')
        display(issues_df)
    
    # Recommendations
    if 'recommendations' in report and report['recommendations']:
        display(Markdown("### Recommendations"))
        for rec in report['recommendations']:
            display(Markdown(f"- {rec}"))
    
except Exception as e:
    print(f"\n Error loading report: {str(e)}")

```
