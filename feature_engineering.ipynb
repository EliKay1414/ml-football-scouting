```python


#  Step 1: Import necessary libraries
import pandas as pd
import numpy as np
import json
from pathlib import Path
import sys
import logging
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.backends.backend_pdf import PdfPages
from datetime import datetime
from sklearn.model_selection import train_test_split

# Configure logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler('feature_engineering.log', encoding='utf-8'),
        logging.StreamHandler()
    ]
)
logger = logging.getLogger(__name__)

# Set display options
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 100)
sns.set_style('whitegrid')
plt.style.use('seaborn-v0_8')
sns.set_palette('colorblind')
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['axes.prop_cycle'] = plt.cycler(color=plt.cm.tab10.colors) 

# Step 2: Import the pipeline module
try:
    # Add your src directory to path
    src_path = Path.cwd().parent / "src"
    sys.path.append(str(src_path))
    
    # Import the pipeline components
    from feature_engineering_pipeline import (
        AdvancedFootballFeatureEngineer,
        ProspectIdentifier,
        ProspectReportGenerator,
        FootballProspectPipeline
    )
    logger.info(" Successfully imported all pipeline components")
    
except Exception as e:
    logger.error(f" Failed to import pipeline components: {e}")
    raise

#  Step 3: Configuration
config = {
    "data": {
        "prospect_max_age": 23,
        "min_potential_diff": 5,
        "min_skill_moves": 2,
        "min_weak_foot": 1,
        "min_physical": 50,
        "min_technical": 50,
        "min_mental": 50,
        "min_skill_progression": 4,
        "target_column": "overall",
        "id_columns": ["sofifa_id", "short_name", "club_name"],
        "test_size": 0.2,
        "random_state": 42
    },
    "features": {
        "base_numerical": ['age', 'height_cm', 'weight_kg', 'overall', 'potential',
                          'value_eur', 'wage_eur', 'weak_foot', 'skill_moves',
                          'international_reputation', 'pace', 'shooting', 'passing',
                          'dribbling', 'defending', 'physic', 'potential_growth',
                          'dev_curve_score', 'skill_progression'],
        "base_categorical": ['preferred_foot', 'player_positions'],
        "composite_features": ['physical_composite', 'technical_composite', 'mental_composite',
                              'defensive_score', 'playmaking_score', 'attacking_score',
                              'goalkeeping_score', 'speed_endurance', 'power_index'],
        "market_metrics": ['value_efficiency', 'wage_ratio', 'contract_score'],
        "raw_attributes_to_keep": [
            'attacking_finishing', 'mentality_vision', 'mentality_composure',
            'defending_standing_tackle', 'power_stamina', 'power_strength',
            'movement_balance', 'skill_ball_control', 'skill_long_passing'
        ]
    }
}
from pathlib import Path

# Step 1: Set up the base data directory
data_dir = Path.cwd().parent / "data"

# Step 2: Define correct paths
train_data_path = data_dir / "splits" / "v1" / "train.csv"
test_data_path = data_dir / "splits" / "v1" / "test.csv"
raw_data_path = data_dir / "processed" / "fifa_players_cleaned.csv"
processed_train_path = data_dir / "processed" / "fifa_players_train.csv"
processed_test_path = data_dir / "processed" / "fifa_players_test.csv"
report_path = data_dir / "reports" / "prospect_report.pdf"


# Step 2: Define file and directory paths
processed_dir = data_dir / "processed"
split_dir = data_dir / "splits" / "v1"
reports_dir = data_dir / "reports"

# Create directories if they don't exist
processed_dir.mkdir(parents=True, exist_ok=True)
split_dir.mkdir(parents=True, exist_ok=True)
reports_dir.mkdir(parents=True, exist_ok=True)

#  Step 3: Define file paths
raw_data_path = processed_dir / "fifa_players_cleaned.csv"
train_data_path = split_dir / "train.csv"         
test_data_path = split_dir / "test.csv"           
report_path = reports_dir / "prospect_report.pdf"

#  Step 4: Load the data for feature engineering
print("Loading training data for feature engineering...")
df_train = pd.read_csv(train_data_path)

print("Training data shape:", df_train.shape)
print("Columns:", df_train.columns.tolist())


#  Step 5: Load Pre-Split Data
try:
    logger.info(f" Loading pre-split training data from {train_data_path}")
    train_df = pd.read_csv(train_data_path)

    logger.info(f" Loading pre-split test data from {processed_test_path}")
    test_df = pd.read_csv(test_data_path)

    logger.info(f" Successfully loaded pre-split data")
    logger.info(f"Train shape: {train_df.shape} | Test shape: {test_df.shape}")

    print("\nSample training data (5 rows):")
    display(train_df.head(5))
    
    print("\nSample test data (5 rows):")
    display(test_df.head(5))

except Exception as e:
    logger.error(f" Failed to load pre-split data: {e}")
    raise

# Step 6: Execute Pipeline with Proper Train-Test Separation
try:
    logger.info(" Initializing football prospect pipeline")
    
    # Initialize pipeline components with config
    feature_engineer = AdvancedFootballFeatureEngineer(fit_mode=True)
    prospect_identifier = ProspectIdentifier(
        min_potential_diff=config["data"]["min_potential_diff"],
        max_age=config["data"]["prospect_max_age"],
        min_skill_moves=config["data"]["min_skill_moves"],
        min_weak_foot=config["data"]["min_weak_foot"],
        min_physical=config["data"]["min_physical"],
        min_technical=config["data"]["min_technical"],
        min_mental=config["data"]["min_mental"],
        min_skill_progression=config["data"]["min_skill_progression"]
    )
    
    # Process training data (without FootballDataCleaner)
    logger.info(" Processing training data")
    engineered_train = feature_engineer.fit_transform(train_df)
    processed_train = prospect_identifier.add_is_prospect(engineered_train)
    
    # Process test data (using fitted transformers)
    logger.info(" Processing test data")
    engineered_test = feature_engineer.transform(test_df)
    processed_test = prospect_identifier.add_is_prospect(engineered_test)
    
    # Select final features
    columns_to_keep = (
        config["features"]["base_numerical"] +
        config["features"]["base_categorical"] +
        config["features"]["composite_features"] +
        config["features"]["market_metrics"] +
        config["data"]["id_columns"] +
        config["features"]["raw_attributes_to_keep"] +
        ['is_prospect']
    )
    
    # Ensure all columns exist in the data
    available_columns = set(processed_train.columns)
    columns_to_keep = [col for col in columns_to_keep if col in available_columns]
    
    train_final = processed_train[columns_to_keep]
    test_final = processed_test[columns_to_keep]
    
    # Save processed data
    train_final.to_csv(processed_train_path, index=False)
    test_final.to_csv(processed_test_path, index=False)
    logger.info(f" Saved processed data to {processed_train_path} and {processed_test_path}")
    
    # Generate report on training data only
    report_gen = ProspectReportGenerator(train_final)
    report_path = report_gen.generate_full_report(report_path)
    logger.info(f" Generated report at {report_path}")
    
    print("\n Pipeline execution complete!")
    print(f"Original features: {len(train_df.columns)}")
    print(f"Processed features: {len(train_final.columns)}")
    print(f"New engineered features: {len(train_final.columns) - len(train_df.columns)}")
    
    print("\nProcessed training data sample (3 rows):")
    display(train_final.head(3))
    
except Exception as e:
    logger.error(f" Pipeline execution failed: {e}")
    raise


#  Step 7: Feature Verification and Analysis (Training Data Only)
print("\n" + "="*50)
print("FEATURE ANALYSIS & VERIFICATION (TRAINING DATA)".center(50))
print("="*50)

# 1. Prospect analysis
if 'is_prospect' in train_final.columns:
    print("\n Prospect Analysis:")
    print(train_final['is_prospect'].value_counts())
    
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    train_final['is_prospect'].value_counts().plot(kind='bar')
    plt.title('Prospect Distribution (Training)')
    
    plt.subplot(1, 2, 2)
    sns.boxplot(x='is_prospect', y='potential_growth', data=train_final)
    plt.title('Potential Growth by Prospect Status')
    plt.tight_layout()
    plt.show()
    
    # Show top prospects
    top_prospects = train_final[train_final['is_prospect'] == 1].sort_values(
        'potential_growth', ascending=False).head(5)
    print("\nTop 5 Prospects:")
    display(top_prospects[['short_name', 'age', 'player_positions', 
                         'overall', 'potential', 'potential_growth']])

# 2. Feature correlations
target_col = config["data"]["target_column"]
if target_col in train_final.columns:
    print(f"\n Top Correlations with {target_col}:")
    numeric_cols = train_final.select_dtypes(include=[np.number]).columns
    correlations = train_final[numeric_cols].corr()[target_col].abs().sort_values(ascending=False)
    display(correlations.head(15))
    
    # Plot top correlations
    top_corrs = correlations.head(10).index
    plt.figure(figsize=(12, 8))
    sns.heatmap(train_final[top_corrs].corr(), annot=True, cmap='coolwarm', fmt=".2f")
    plt.title('Top Feature Correlations (Training)')
    plt.show()

# 3. Composite score distributions
composite_features = [f for f in config["features"]["composite_features"] if f in train_final.columns]
if composite_features:
    print("\n Composite Scores Analysis:")
    
    plt.figure(figsize=(12, 8))
    for i, score in enumerate(composite_features[:6], 1):  # limit to first 6 features
        plt.subplot(2, 3, i)
        sns.histplot(train_final[score], kde=True, bins=30)
        plt.title(f'Distribution of {score}')

    plt.tight_layout()
    plt.show()
    
    # Compare with overall rating
    plt.figure(figsize=(10, 6))
    for score in composite_features:
        sns.regplot(x=score, y=target_col, data=train_final, 
                   scatter_kws={'alpha':0.3}, label=score)
    plt.legend()
    plt.title('Composite Scores vs Overall Rating')
    plt.show()

# 4. Position analysis
if 'player_positions' in train_final.columns:
    print("\n Position Distribution:")
    position_counts = train_final['player_positions'].value_counts().head(10)
    display(position_counts)
    
    plt.figure(figsize=(10, 5))
    position_counts.plot(kind='bar')
    plt.title('Top 10 Player Positions (Training)')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()

# 5. Market metrics analysis
market_metrics = [f for f in config["features"]["market_metrics"] if f in train_final.columns]
if market_metrics:
    print("\n Market Metrics Analysis:")
    
    plt.figure(figsize=(15, 5))
    for i, metric in enumerate(market_metrics, 1):
        plt.subplot(1, 3, i)
        sns.histplot(train_final[metric], kde=True, bins=30)
        plt.title(f'Distribution of {metric}')
    plt.tight_layout()
    plt.show()

#  Step 8: Data Quality Checks
print("\n" + "="*50)
print("DATA QUALITY CHECKS".center(50))
print("="*50)

# Missing values
print("\n Missing Values in Training Data:")
missing_values = train_final.isnull().sum()
print(missing_values[missing_values > 0].sort_values(ascending=False))

# Data types
print("\n Data Types:")
print(train_final.dtypes.value_counts())

# Numeric features summary
print("\n Numeric Features Summary:")
display(train_final.select_dtypes(include=[np.number]).describe().loc[['mean', 'std', 'min', 'max']].T.head(10))

# Verify test set transformation
print("\n Test Set Verification:")
print(f"Test set shape: {test_final.shape}")
print("Columns match:", set(train_final.columns) == set(test_final.columns))
if 'potential_growth' in train_final.columns and 'potential_growth' in test_final.columns:
    print("\nPotential Growth Stats Comparison:")
    print("Training - Mean:", train_final['potential_growth'].mean())
    print("Test - Mean:", test_final['potential_growth'].mean())

# Save the configuration for reference
config_path = data_dir / "processed" / "feature_engineering_config.json"
with open(config_path, 'w') as f:
    json.dump(config, f, indent=2)
print(f"\nConfiguration saved to {config_path}")

# Save the list of features for modeling
features_path = data_dir / "processed" / "modeling_features.txt"
with open(features_path, 'w') as f:
    f.write("\n".join(columns_to_keep))
print(f"Feature list saved to {features_path}")

logger.info(" Feature engineering pipeline completed successfully with proper train-test separation")
```
