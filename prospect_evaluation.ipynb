```python
# %% [markdown]
"""
# Football Prospect Scouting - Model Evaluation Notebook
This notebook executes the model evaluation pipeline for prospect identification models.
"""

# %%
# Setup environment
import sys
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import json
import warnings
from IPython.display import display, Markdown
warnings.filterwarnings('ignore')

# Set up project paths
project_root = Path(r"C:\Users\Nanaba\Desktop\football_player_scouting_ml")
sys.path.insert(0, str(project_root / "src"))

# Import your evaluation module
from evaluation import ProspectModelEvaluator

# Visualization setup
plt.style.use('seaborn-v0_8')
sns.set_theme(style="whitegrid", palette="viridis")
%matplotlib inline

# %%
# Configuration
config = {
    "paths": {
        "models_dir": str(project_root / "models" / "prospect_models_v1"),
        "test_data": str(project_root / "data" / "features" / "test_features.csv"),
        "output_dir": str(project_root / "evaluation_results")
    },
    "metrics": {
        "threshold": 0.5,
        "positive_class": 1
    }
}

# Define path to save config file
config_path = project_root / "config_eval.json"

# Convert Path objects to strings for JSON serialization
config_serializable = config.copy()
config_serializable["paths"] = {k: str(v) for k, v in config["paths"].items()}

# Save config
with open(config_path, 'w') as f:
    json.dump(config_serializable, f, indent=2)



# %%
# Initialize the evaluator
evaluator = ProspectModelEvaluator(str(config_path))

# %%
# Load models
print("Loading trained models...")
models = evaluator.load_models()
print(f"\nLoaded {len(models)} models:")
for model_name in models.keys():
    print(f"- {model_name}")

# %%
# Load test data
print("\nLoading test data...")
X_test, y_test = evaluator.load_test_data()

print(f"\nTest set: {X_test.shape[0]} samples")
print("\nClass distribution:")
print(pd.Series(y_test).value_counts(normalize=True))

# %%
# Evaluate models
print("\nEvaluating models on test set...")
results = evaluator.evaluate_models(X_test, y_test)

# Show evaluation results
print("\nEvaluation Results:")
results_df = pd.DataFrame.from_dict(results, orient='index')
metrics_to_show = ['accuracy', 'roc_auc', 'average_precision']
display(results_df[metrics_to_show].sort_values('roc_auc', ascending=False))

# %%
# Generate detailed classification reports
print("\nDetailed Classification Reports:")
for model_name, metrics in results.items():
    if 'classification_report' in metrics:
        print(f"\n{model_name}:")
        display(pd.DataFrame(metrics['classification_report']).transpose())

# %%
# Generate visualizations
print("\nGenerating evaluation visualizations...")
evaluator.generate_visualizations()

# Display ROC Curves
print("\nROC Curves:")
roc_img = plt.imread(Path(config["paths"]["output_dir"]) / "roc_curves.png")
plt.figure(figsize=(10, 6))
plt.imshow(roc_img)
plt.axis('off')
plt.show()

# Display Precision-Recall Curves
print("\nPrecision-Recall Curves:")
pr_img = plt.imread(Path(config["paths"]["output_dir"]) / "precision_recall_curves.png")
plt.figure(figsize=(10, 6))
plt.imshow(pr_img)
plt.axis('off')
plt.show()

# %%
# Display confusion matrices for each model
print("\nConfusion Matrices:")
for model_name in models.keys():
    try:
        img_path = Path(config["paths"]["output_dir"]) / f"confusion_matrix_{model_name}.png"
        if img_path.exists():
            print(f"\n{model_name}:")
            img = plt.imread(img_path)
            plt.figure(figsize=(6, 6))
            plt.imshow(img)
            plt.axis('off')
            plt.show()
    except Exception as e:
        print(f"Could not display confusion matrix for {model_name}: {str(e)}")

# %%
# Display feature importance plots
print("\nFeature Importance Plots:")
for model_name in ['XGBoost', 'RandomForest', 'LightGBM']:  # Models that support feature importance
    try:
        img_path = Path(config["paths"]["output_dir"]) / f"feature_importance_{model_name}.png"
        if img_path.exists():
            print(f"\n{model_name}:")
            img = plt.imread(img_path)
            plt.figure(figsize=(10, 8))
            plt.imshow(img)
            plt.axis('off')
            plt.show()
    except Exception as e:
        print(f"Could not display feature importance for {model_name}: {str(e)}")

# %%
# Save results
print("\nSaving evaluation results...")
evaluator.save_results()

print("\nEvaluation results saved to:", config["paths"]["output_dir"])

# %%
print("\nEvaluation pipeline completed successfully!")
```
