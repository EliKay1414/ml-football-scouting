```python
# %% [markdown]
# Setup environment
import sys
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import json
import warnings
from IPython.display import display, Markdown
warnings.filterwarnings('ignore')

# Set up project paths
project_root = Path(r"C:\Users\Nanaba\Desktop\football_player_scouting_ml")
sys.path.insert(0, str(project_root / "src"))

# Import evaluation module
from evaluation_deployment import (
    ProspectModelEvaluator,
    rank_models_by_each_metric,
    rank_models_by_weighted_score
)


# Visualization setup
plt.style.use('seaborn-v0_8')
sns.set_theme(style="whitegrid", palette="viridis")
%matplotlib inline

import json
from pathlib import Path

# Define your project root folder
project_root = Path("C:/Users/Nanaba/Desktop/football_player_scouting_ml")

# Configuration dictionary
config = {
     "paths": {
        "models_dir": project_root / "models" / "prospect_models_v1",
        "test_features": project_root / "data" / "leakproof_features" / "processed_test.feather",
        "selected_features": project_root / "data" / "leakproof_features" / "selected_features.pkl",
        "test_labels": project_root / "data" / "leakproof_features" / "test_labels.csv",
        "new_data": project_root / "data" / "new_players.csv",
        "output_dir": project_root / "evaluation_results",
        "deployment_dir": project_root / "models" / "deployment"
    },

    "metrics": {
        "threshold": 0.5,
        "positive_class": 1,
        "top_n_features": 15,
        "min_recall": 0.6,     # Minimum recall requirement
        "min_precision": 0.6  # Minimum precision requirement
    }
}

# Save config to JSON file (convert Path objects to strings)
config_path = project_root / "config_eval.json"
with open(config_path, 'w') as f:
    json.dump({
        "paths": {k: str(v) for k, v in config["paths"].items()},
        "metrics": config["metrics"]
    }, f, indent=2)

print(f"Configuration saved to {config_path}")


# %%
# Initialize the evaluator
print("Initializing evaluator with Pareto frontier support...")
evaluator = ProspectModelEvaluator(str(config_path))

# %%
# 1. Load models and test data
print("Loading trained models...")
models = evaluator.load_models()
print(f"\nLoaded {len(models)} models:")
for model_name in models.keys():
    print(f"- {model_name}")

print("\nLoading test data...")
X_test, y_test = evaluator.load_test_data()

print(f"\nTest set: {len(X_test)} samples")
print("Class distribution:")
print(pd.Series(y_test).value_counts(normalize=True))

# %%
# 2. Evaluate models and compare performance
print("\nEvaluating models on test set...")
results = evaluator.evaluate_models(X_test, y_test)

print("\nModel Comparison:")
comparison_df = evaluator.compare_models()
display(comparison_df.style.background_gradient(
    subset=['roc_auc', 'f1_score', 'recall', 'precision'], 
    cmap='YlGn'

))
rank_models_by_each_metric(evaluator)
rank_models_by_weighted_score(evaluator)

# %%
# 3. Pareto Frontier Analysis 
print("\n=== Pareto Frontier Analysis ===")
pareto_df = evaluator.get_pareto_frontier()
print("\nPareto-optimal models (non-dominated solutions):")
display(pareto_df[['model', 'roc_auc', 'f1_score', 'recall', 'precision']])

# Visualize
print("\nGenerating Pareto frontier plots...")
evaluator.plot_pareto_frontier(x_metric="recall", y_metric="precision")
evaluator.plot_pareto_frontier(x_metric="recall", y_metric="f1_score")
# After evaluating models
pareto_df = evaluator.get_pareto_frontier()
print("Pareto-efficient models:")
display(pareto_df[pareto_df["is_pareto"]])

# Visualize
evaluator.plot_pareto_frontier(x_metric="recall", y_metric="precision")

# Select best model with recall constraint
best_model = evaluator.select_best_model(
    priority_metrics=["f1_score", "roc_auc", "recall", "precision", "accuracy"],
    weights={
        "f1_score": 0.3,
        "roc_auc": 0.3,
        "recall": 0.2,
        "precision": 0.2,
        "accuracy": 0.0
    },
    min_recall=0.6,
    min_precision=0.6
)


# Show precision-recall frontier
plt.figure(figsize=(12, 6))
img = plt.imread(Path(config["paths"]["output_dir"]) / "pareto_recall_vs_precision.png")
plt.imshow(img)
plt.axis('off')
plt.title('Precision-Recall Pareto Frontier')
plt.show()

# %%
# 4. Select Best Model with Business Constraints 
print("\n=== Model Selection with Constraints ===")
print(f"Business requirements:\n- Minimum Recall: {config['metrics']['min_recall']}\n- Minimum Precision: {config['metrics']['min_precision']}")

best_model = evaluator.select_best_model(
    priority_metrics=["roc_auc", "f1_score", "recall", "accuracy", "precision"],
    weights={
        "roc_auc": 0.3,
        "f1_score": 0.3,
        "recall": 0.2,
        "accuracy": 0.0,
        "precision": 0.2
    },
    min_recall=0.6,
    min_precision=0.6
)

best_model_name = evaluator.best_model["name"]
print(f"\nSelected best model: {best_model_name}")

best_metrics = evaluator.best_model["metrics"]
print("\nModel Performance:")
print(f"- ROC AUC: {best_metrics['roc_auc']:.3f}")
print(f"- F1 Score: {best_metrics['classification_report']['1']['f1-score']:.3f}")
print(f"- Recall: {best_metrics['classification_report']['1']['recall']:.3f}")
print(f"- Precision: {best_metrics['classification_report']['1']['precision']:.3f}")

# Save the model
best_model_path = evaluator.save_best_model()
print(f"\nSaved best model to: {best_model_path}")

# %%
# Load test labels and align them with features
label_path = config["paths"]["test_labels"]
test_labels = pd.read_csv(label_path)

# Ensure proper column name
test_labels.rename(columns={test_labels.columns[0]: "is_prospect"}, inplace=True)

# Align y_test to X_test by index
y_test = test_labels.loc[X_test.index, "is_prospect"]

# Create test set with both features and label
test_data_with_labels = X_test.copy()
test_data_with_labels["is_prospect"] = y_test

# Assign to evaluator
evaluator.test_data = test_data_with_labels


# Assign to evaluator for downstream steps
evaluator.test_data = test_data_with_labels


# Rename label column if needed
test_labels.rename(columns={test_labels.columns[0]: "is_prospect"}, inplace=True)

# Combine into one DataFrame
test_data_with_labels = X_test.copy()
test_data_with_labels["is_prospect"] = test_labels["is_prospect"]

print("\n=== Optimal Threshold Calculation ===")
thresholds = evaluator.find_optimal_thresholds(
    min_precision=config['metrics']['min_precision'],
    min_recall=config['metrics']['min_recall']
)

print("\nRecommended thresholds:")
threshold_df = pd.DataFrame.from_dict(thresholds, orient='index')
display(threshold_df.style.format({
    'threshold': '{:.3f}',
    'precision': '{:.1%}',
    'recall': '{:.1%}'
}))

if best_model_name in thresholds and thresholds[best_model_name]:
    optimal_threshold = thresholds[best_model_name]["threshold"]
else:
    optimal_threshold = 0.5  # Fallback
print(f"\nUsing threshold {optimal_threshold:.3f} for {best_model_name}")

# %%
# 7. Generate Full Report
print("\nGenerating comprehensive evaluation report...")
evaluator.generate_report()

# Display key visualizations
print("\n=== Key Evaluation Visualizations ===")

# ROC Curves
roc_img = plt.imread(Path(config["paths"]["output_dir"]) / "roc_curves.png")
plt.figure(figsize=(10, 6))
plt.imshow(roc_img)
plt.axis('off')
plt.title('ROC Curves')
plt.show()

# Precision-Recall Curves
pr_img = plt.imread(Path(config["paths"]["output_dir"]) / "precision_recall_curves.png")
plt.figure(figsize=(10, 6))
plt.imshow(pr_img)
plt.axis('off')
plt.title('Precision-Recall Curves')
plt.show()

# Calibration Plot
cal_img = plt.imread(Path(config["paths"]["output_dir"]) / "calibration_plots.png")
plt.figure(figsize=(10, 6))
plt.imshow(cal_img)
plt.axis('off')
plt.title('Model Calibration')
plt.show()
# Display confusion matrix
print("\n=== Confusion Matrices ===")
cm_img = plt.imread(Path(config["paths"]["output_dir"]) / "confusion_matrices.png")
plt.figure(figsize=(12, 8))
plt.imshow(cm_img)
plt.axis('off')
plt.title('Model Confusion Matrices')
plt.show()

# %%
print("\n Evaluation pipeline completed successfully!")
```
